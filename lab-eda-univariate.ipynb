```python
# Remove rows where price is missing or non-numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['price'])

# Calculate mean, median, and mode
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]  # mode() returns a series, so we take the first value

print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")
```

```python
# Calculate variance, standard deviation, range, and IQR
price_variance = df['price'].var()
price_std_dev = df['price'].std()
price_range = df['price'].max() - df['price'].min()
price_iqr = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Variance: {price_variance}")
print(f"Standard Deviation: {price_std_dev}")
print(f"Range: {price_range}")
print(f"IQR: {price_iqr}")
```

```python
# Plot histogram for price distribution
plt.figure(figsize=(8,6))
plt.hist(df['price'], bins=50, color='blue', edgecolor='black')
plt.title('price Distribution of Products')
plt.xlabel('price')
plt.ylabel('frequency')
plt.show()
```

```python
# Box plot for price spread and outliers
plt.figure(figsize=(8,6))
plt.boxplot(df['price'], vert=False)
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Remove rows where 'stars' is missing or non-numeric
df['stars'] = pd.to_numeric(df['stars'], errors='coerce')
df = df.dropna(subset=['stars'])

# Calculate mean, median, and mode
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

print(f"Mean Rating: {mean_rating}")
print(f"Median Rating: {median_rating}")
print(f"Mode Rating: {mode_rating}")

```

```python
# Calculate variance, standard deviation, and IQR for ratings
rating_variance = df['stars'].var()
rating_std_dev = df['stars'].std()
rating_iqr = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

print(f"Variance: {rating_variance}")
print(f"Standard Deviation: {rating_std_dev}")
print(f"IQR: {rating_iqr}")
```

```python
# Calculate skewness and kurtosis for product ratings
skewness = df['stars'].skew()
kurtosis = df['stars'].kurtosis()

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")
```

```python
# Plot histogram for ratings
plt.figure(figsize=(8,6))
plt.hist(df['stars'], bins=10, color='green', edgecolor='black')
plt.title('Rating Distribution of Products')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

In this analysis, we examined product ratings on Amazon UK to uncover patterns in customer feedback. Understanding how customers rate products can provide valuable insights into customer satisfaction and product performance.


Measures of Central Tendency

Mean Rating: The average customer rating is 2.15, which suggests that on average, customers rate products relatively low on a scale of 0 to 5.

Median Rating: The median rating is 0.0, meaning that at least half of the products are rated 0. This is surprising and indicates that a significant portion of products may not be receiving positive ratings at all.

Mode Rating: The most frequent rating is 0.0, further confirming that a large number of products have low or no ratings.

Interpretation: The fact that both the median and mode are 0.0 while the mean is slightly higher at 2.15 suggests that while many products are rated very low, a smaller number of products might be pulling the mean up with higher ratings. This implies that a substantial portion of products may not be meeting customer expectations.



Measures of Dispersion

Variance: The variance of 4.82 indicates a wide spread in the ratings, suggesting inconsistency in customer feedback.

Standard Deviation: The standard deviation of 2.19 reinforces the idea that ratings vary significantly from the mean.

Interpretation: High variance and standard deviation indicate that there’s a lot of variability in how customers rate products. While some products might be highly rated, others receive very low or no ratings, which reflects inconsistent product satisfaction.

Interquartile Range (IQR): The IQR of 4.4 shows that the middle 50% of ratings are spread across a large range of values.

Interpretation: A large IQR indicates that there is significant variation in the ratings, with many ratings being quite different from each other.



Shape of the Distribution

Skewness: The skewness value of 0.081 indicates that the distribution is nearly symmetrical, meaning ratings don’t lean strongly towards either high or low values.

Interpretation: A skewness close to 0 suggests that while a significant number of products have low ratings, the distribution is balanced, with higher ratings also present in the dataset.

Kurtosis: The kurtosis of -1.93 indicates a distribution with fewer extreme outliers than a normal distribution.

Interpretation: The negative kurtosis suggests that most ratings are concentrated around the middle, with fewer extreme high or low ratings than expected. However, this is unusual given that the median and mode are 0.0, which could suggest data issues or specific products receiving many zero ratings.
